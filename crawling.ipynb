```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
def parse_value(value_str):
    # 예: '50.00 mil. €' → 50000000
    try:
        num = float(value_str.replace("mil. €", "").strip())
        return int(num * 1_000_000)
    except:
        return 0  # 예외 처리
```

```python

# Selenium 드라이버 설정 (Chrome 사용)
driver = webdriver.Chrome()

def accept_cookie(driver):
    try:
        WebDriverWait(driver, 3).until(
            EC.frame_to_be_available_and_switch_to_it((By.ID, "sp_message_iframe_953770"))
        )
        WebDriverWait(driver, 3).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "button.accept-all"))
        ).click()
        driver.switch_to.default_content()
        print("쿠키 수락 완료")
    except:
        print("쿠키 팝업 없음 또는 이미 수락됨")
player_list = []
idx = 1
for page in range(1, 5):
    url = f"https://www.transfermarkt.co.kr/spieler-statistik/wertvollstespieler/marktwertetop?page={page}"
    driver.get(url)
    print(f"\n📄 페이지 {page} 접근 완료")
    time.sleep(2)
    accept_cookie(driver)

    print(f"\n 페이지 {page} 접근 완료")
    players = driver.find_elements(By.CSS_SELECTOR, "table.items tbody tr")
    for p in players:
        try:
            # 이름
            name = p.find_element(By.CSS_SELECTOR, "td.hauptlink a").text.strip()

            # 포지션 (두 번째 줄 텍스트)
            position = p.find_element(By.CSS_SELECTOR, "table.inline-table tr:nth-child(2) td").text.strip()

            # 나이
            age = p.find_element(By.CSS_SELECTOR, "td.zentriert:nth-of-type(3)").text.strip()

            # 팀
            team = p.find_element(By.CSS_SELECTOR, "td.zentriert:nth-of-type(5) img").get_attribute("title")

            # 시장 가치
            value = p.find_element(By.CSS_SELECTOR, "td.rechts.hauptlink").text.strip()
            value_eur = parse_value(value)
            value_krw = int(value_eur * 1550)
            value_uk = round(value_krw / 100000000)  # 억 단위로 변환

            info = f"{idx}위. {name}선수는 {position}포지션이며 나이는 {age}살 입니다. {team}에서 뛰고 있으며 현재 몸값은 {value_uk}억 원입니다."
            # 리스트에 저장 직전 확인
            print("추출된 정보:", info)
            player_list.append(info)
            idx += 1
        except Exception as e:
            continue
driver.quit()
```

```python
with open("source/players.txt", "w", encoding="utf-8") as f:
    f.write("2025년 전세계 축구선수 몸값 현황\n")
    f.write("출처 : 트렌스퍼마켓\n\n")
    for player in player_list:
        f.write(player + "\n")

print("source/players.txt 저장 완료!")
```
